```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Closing Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Define the stock symbol and time range for GameStop
stock_symbol_gme = "GME"
start_date = "2021-01-01"
end_date = "2022-01-01"

# Fetch GameStop stock data
gme_stock_data = yf.download(stock_symbol_gme, start=start_date, end=end_date)

# Reset the index for consistency
gme_stock_data.reset_index(inplace=True)

# Call the make_graph function
make_graph(gme_stock_data, title='GameStop Stock Price Graph')
```
