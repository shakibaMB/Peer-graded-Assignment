```python
import matplotlib.pyplot as plt
import yfinance as yf

```

```python
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Closing Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Define the stock symbol and time range for Tesla
stock_symbol_tsla = "TSLA"
start_date = "2021-01-01"
end_date = "2022-01-01"

# Fetch Tesla stock data
tesla_stock_data = yf.download(stock_symbol_tsla, start=start_date, end=end_date)

# Reset the index for consistency
tesla_stock_data.reset_index(inplace=True)

# Call the make_graph function
make_graph(tesla_stock_data, title='Tesla Stock Price Graph')
```
